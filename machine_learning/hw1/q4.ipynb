```python
# Hasan F. Durkaya 504241526
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data
X = np.array([69, 74, 68, 70, 72, 67, 66, 70, 76, 68, 72, 79, 74, 67, 66, 71, 74, 75, 75, 76])
Y = np.array([153, 175, 155, 135, 172, 150, 115, 137, 200, 130, 140, 265, 185, 112, 140, 150, 165, 185, 210, 220])
```

```python
# a) Mean, median, and mode calculations
mean_X = np.mean(X)
median_X = np.median(X)
mode_X = max(set(X), key=list(X).count)

print(f"Mean of X: {mean_X}")
print(f"Median of X: {median_X}")
print(f"Mode of X: {mode_X}")
```

```python
# b) Variance oıf X
variance_Y = np.sum((Y - np.mean(Y))**2) / (len(Y) - 1)  
print(f"Variance of Y: {variance_Y}")
```

```python
# c) Plotting the normal distribution for X
# Calculating mean and standard deviation
std = np.std(X, ddof=1)

# Calculating the probability density function (pdf) for the normal distribution
pdf = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((X - mean_X) / std)**2)

# Plotting the normal distribution
plt.figure(figsize=(8, 5))
plt.plot(X, pdf, label="Normal Distribution", color="blue")
plt.title("Normal Distribution of X (Ages)")
plt.xlabel("Age")
plt.ylabel("Probability Density")
plt.grid(alpha=0.3)
plt.legend()
plt.show()

```

```python
# d) Probability of observing age ≥ 80
def covariance(x, y):
    return np.sum((x - np.mean(x)) * (y - np.mean(y))) / (len(x) - 1)


probability = (np.sum(X > 80) / len(X)) * 100
print(f"Probability of observing age ≥ 80: {probability:.4f}")

# e) 2D mean and covariance matrix
mean_XY = np.array([np.mean(X), np.mean(Y)])
cov_matrix = np.array([
    [covariance(X, X), covariance(X, Y)],
    [covariance(Y, X), covariance(Y, Y)]
])

print(f"2D Mean Vector:\n{mean_XY}")
print(f"Covariance Matrix:\n{cov_matrix}")
```

```python
# f) Correlation between age and weight
correlation = cov_matrix[0, 1] / (np.sqrt(cov_matrix[0, 0] * cov_matrix[1, 1]))
print(f"Correlation between X (Age) and Y (Weight): {correlation:.4f}")

# g) Scatterplot
plt.figure(figsize=(8, 5))
plt.scatter(X, Y, color="blue", alpha=0.7)
plt.title("Scatterplot: Age vs Weight")
plt.xlabel("Age (X)")
plt.ylabel("Weight (Y)")
plt.grid(alpha=0.3)
plt.show()
```
