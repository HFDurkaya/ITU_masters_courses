```python
# Hasan F. Durkaya 504241526
import numpy as np
import matplotlib.pyplot as plt

"""
Script visualizes the Laplace distribution using matplotlib.
"""

def laplace_distribution(x):
    return 0.5 * np.exp(-np.abs(x))

x = np.linspace(-15, 15, 1000)
y = laplace_distribution(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, color='red', label='Laplace Distribution')
plt.title("Laplace Distribution")
plt.xlabel("x")
plt.ylabel("p(x)")
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
